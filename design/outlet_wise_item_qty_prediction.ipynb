## Importing data from database into pandas dataframes

```python
query = "SELECT Outlet_ID,ITEM_CODE,Qty FROM handheld_sales_master;"
order_df = pd.read_sql(query, cnxn)
print(order_df)

query1 = "SELECT Outletid,OutletName FROM Outlet_Master;"
outlet_df = pd.read_sql(query1, cnxn)
print(outlet_df)

query3 = "SELECT item_code,item_descn,nob FROM mast_item where finished_product='Y';"
item_df = pd.read_sql(query3, cnxn)
print(item_df)

query4 = "SELECT OutletId,ItemId,SQty FROM Stock_Trns ;"
cs_df = pd.read_sql(query4, cnxn)
print(cs_df)
```

## Renaming columns

```python
order_df = order_df.rename(columns={'Outlet_ID':'Outletid','ITEM_CODE':'item_code'}) 
order_df
```

```python
cs_df = cs_df.rename(columns={'OutletId':'Outletid','ItemId':'item_code'}) 
cs_df
```

## Merging order and item dataframes

```python
order_item = pd.merge(order_df,item_df,on='item_code',how='inner')
order_item
```

## Merging order, item and outlet dataframes

```python
order_item_outlet = pd.merge(order_item,outlet_df,on='Outletid',how='inner')
order_item_outlet  
```

## Merging outlet_current_stock and outlet dataframes

```python
cs_outlet = pd.merge(cs_df,outlet_df,on='Outletid',how='inner')
cs_outlet
```

## Merging outlet_current_stock, outlet and item dataframes

```python
cs_outlet_item = pd.merge(cs_outlet,item_df,on='item_code',how='inner')
cs_outlet_item
```

## Orders dataframe for a single outlet

```python
for_each_outlet = order_item_outlet.loc[order_item_outlet['OutletName'] == 'BARSA SWEETS']
for_each_outlet = for_each_outlet.reset_index(drop=True)
for_each_outlet
```

## Converting order quantity from case to bottles 

```python
nob_lists=[]
for row in for_each_outlet.itertuples():
    sql = """\
    DECLARE @RC int;
    EXEC @RC = [CaseToBtl] @QTY = ?, @NOB = ?;
    SELECT @RC AS rc;
    """
    values = (row.Qty,row.nob)
    cursor.execute(sql, values)
    rc = cursor.fetchval()
    nob_lists.append(rc)
    
for_each_outlet.insert(3, "Qty_in_nob", nob_lists, True)
for_each_outlet 
```

## Current stock dataframe for each outlet

```python
cs_for_each_outlet = cs_outlet_item.loc[cs_outlet_item['OutletName'] == 'BARSA SWEETS']
cs_for_each_outlet = cs_for_each_outlet.reset_index(drop=True)
cs_for_each_outlet
```

## Converting current stock quantity from case to bottle

```python
nob_list=[]
for row in cs_for_each_outlet.itertuples():
    sql = """\
    DECLARE @RC int;
    EXEC @RC = [CaseToBtl] @QTY = ?, @NOB = ?;
    SELECT @RC AS rc;
    """
    values = (row.SQty,row.nob)
    cursor.execute(sql, values)
    rc = cursor.fetchval()
    nob_list.append(rc)
    
cs_for_each_outlet.insert(3, "SQty_in_nob", nob_list, True)
cs_for_each_outlet 
```

## For each item calculating quantity

```python
item_qty_count = (for_each_outlet.
     groupby(['item_descn']).
     mean().
     reset_index()
     [['item_descn', 'Qty', 'Qty_in_nob']]
    )
item_qty_count['1.5xQty'] = item_qty_count['Qty_in_nob']*1.5
item_qty_count
```

## Calculating Suggested Quantity

```python
for i in item_qty_count['item_descn']:
    if (cs_for_each_outlet['item_descn'] == i).any() :
        item_qty_count['Suggested_Qty_in_nob'] = (item_qty_count['1.5xQty'] - cs_for_each_outlet['SQty_in_nob'])
            
item_qty_count
```

## Recommendations

```python
# if 'Suggested_Qty' not in item_qty_count:
#     print("Nothing to recommend")
# else:    
recom_item_qty = item_qty_count.loc[item_qty_count['Suggested_Qty_in_nob']>0].round(2)
recom_item_qty = recom_item_qty.reset_index(drop=True)
recom_item_qty = pd.merge(recom_item_qty,item_df,on='item_descn',how='inner')

recom_item_qty
```

## Converting quantity from bottle to case

```python
nob_lt=[]
for row in recom_item_qty.itertuples():
    sql = """\
    DECLARE @RC float;
    EXEC @RC = [QtyToBtlCase] @QTY = ?, @NOB = ?;
    SELECT @RC AS rc;
    """
    values = (row.Suggested_Qty_in_nob,row.nob)
    cursor.execute(sql, values)
    rc = cursor.fetchval()
    rc
    nob_lt.append(rc)
  
recom_item_qty.insert(5, "Suggested_Qty_in_case", nob_lt, True)
recom_item_qty['Qty_in_nob']=recom_item_qty['Qty_in_nob'].apply(np.ceil)
recom_item_qty['1.5xQty']=recom_item_qty['1.5xQty'].apply(np.ceil)
recom_item_qty['Suggested_Qty_in_nob']=recom_item_qty['Suggested_Qty_in_nob'].apply(np.ceil)
recom_item_qty = recom_item_qty.rename(columns={'Qty':'Average_Qty','Qty_in_nob':'Average_Qty_in_nob','1.5xQty':'Average_Qty_in_nobx1.5'}) 
```
